```python
def g_xi(i):
    gxi = np.sum(y_train * np.array(Kernals[i]) * alpha.reshape(1,-1)) + b
    return gxi

def E_i(i):
    Ei = g_xi(i) - y_train[i]
    return Ei

def selectJrand(i, m):  # 该辅助函数用于在某个区间范围内随机选择一个整数
    j=i                 # m是所有alpha的数目，i是第一个alpha的下标
    while (j==i):
        j = int(random.uniform(0,m))  # random.uniform(0,m)用于生成指定范围内的随机浮点数
    return j

def SimpleSMO(X_train, y_train, Maxiter, C, tolerance):
    X_train = np.mat(X_train)
    y_train = np.mat(y_train).transpose()
    m, n = X_train.shape
    alpha = np.zeros([m, 1])
    b = 0
    iterm = 0
    while iterm < Maxiter:
        change = 0
        for i in range(alpha.size):
            yi = y_train[i]
            Ei = E_i(i)
            ## 违反KKT条件的alpha, 找第一个alpha
            if (yi*Ei < -tolerance and alpha[i][0] < C) or (yi*Ei > tolerance and alpha[i][0] > 0):
                j = selectJrand(i,m)   ## 找第二个alpha，找到与i不同的j
                Ej = E_i(j)
                yj = y_train[j]
                alpha_i_old = alpha[i][0].copy()
                alpha_j_old = alpha[j][0].copy()
                ## 剪辑条件
                if yi != yj:
                    L = max(0, alpha_j_old - alpha_i_old)
                    H = min(C, C - alpha_i_old + alpha_j_old)
                else:
                    L = max(0, alpha_i_old + alpha_j_old - C)
                    H = min(C, alpha_i_old + alpha_j_old)     
                if L == H:
#                     print("L=H")
                    continue      #终止对这一段alpha[i][0]和alpha[j][0]的优化
                ## 未剪辑的解
                eta = Kernals[i, i] + Kernals[j, j] - 2*Kernals[i, j]
                alpha_j_new_unc = alpha_j_old + (yj*(Ei - Ej))/eta
                ##剪辑后alpha[j]的解
                alpha[j][0] = np.clip(alpha_j_new_unc, L, H)
                if abs(alpha[j][0] - alpha_j_old)<0.00001:
#                     print("j not moving")
                    continue      #终止对这一段alpha[i]和alpha[j]的优化
                ##以此来更新alpha[i]的解
                alpha[i][0] = alpha_i_old + yi*yj*(alpha_j_old - alpha[j][0])
                ## 更新b
                bi_new = -Ei-yi*Kernals[i, i]*(alpha[i][0]-alpha_i_old)-yj*Kernals[j, i]*(alpha[j][0]-alpha_j_old)+b
                bj_new = -Ej-yi*Kernals[i, j]*(alpha[i][0]-alpha_i_old)-yj*Kernals[j, j]*(alpha[j][0]-alpha_j_old)+b
                ##设置b_new的值
                if (alpha[i][0] > 0 and alpha[i][0] < C):
                    b = bi_new
                elif (alpha[j][0] > 0 and alpha[j][0] < C):
                    b = bj_new
                else:
                    b = (bi_new + bj_new)/2
                change += 1
        if change == 0:
            iterm += 1
        else:
            iterm = 0
    return alpha, b
```

```python
data, label = datasets.make_classification(n_samples=125,n_features=2,n_classes=2,n_redundant=0,random_state=202289)
label = np.where(label == 0, -1, 1)
X_train, X_test, y_train, y_test = train_test_split(data, label, test_size = 0.2, random_state=202289) 
# data, label = datasets.make_hastie_10_2(n_samples=100, random_state=202288)   #100条数据，10个维度
print(X_train.shape)
print(y_train.shape)
```

```python
class SimpleSMO(object):
    
    def __init__(self,X,y,b,C,tolerance,MaxIter):
        
        
        self.X = X
        self.y = y
        self.b = b
        self.C = C
        self.MaxIter = MaxIter
        self.tolerance = tolerance
        self.alpha = np.zeros((self.X.shape[0],1))
        
        
    def g(self,x_i):
        #计算对输入x_i的预测值
        #此处必须reshape(-1,1)，否则就会变成100*100矩阵
        predicted_y_i = np.sum(self.alpha * self.y.reshape(-1, 1) * (np.dot(self.X,x_i.T).reshape(-1,1))) + self.b
        return predicted_y_i

    def Error(self,x_i,y_i):
        #计算预测值与输入值的误差 
        return self.g(x_i) - y_i
    
    def SelectJ(self,i):
        #简化版SMO:随机选择第二个优化变量j，并使其不等于第一个i
        j = i
        while (j==i):
            j = int(random.uniform(0,self.X.shape[0]))
        return j
    
    def Kernal(self,m,n):
        #定义核函数，用于计算Kij，本例中Kij = X[i].*X[j]
        return self.X[m].dot(self.X[n].T)
        
    
    def Optimization(self):
        iter = 0 
        while (iter < self.MaxIter):
            alphaPairsChanged = 0
            for i in range(self.alpha.size):
                E_i = self.Error(self.X[i],self.y[i])
                if (y[i] * E_i < -self.tolerance and self.alpha[i] < self.C ) \
                or (y[i] * E_i > self.tolerance and self.alpha[i] > 0 ):
                    j = self.SelectJ(i)
                    E_j = self.Error(self.X[j],self.y[j])
                    alpha_i_old = self.alpha[i].copy()
                    alpha_j_old = self.alpha[j].copy()
                    if (self.y[i] != self.y[j]):
                        L = max(0,alpha_j_old-alpha_i_old)
                        H = min(self.C,self.C + alpha_j_old - alpha_i_old)
                    else:
                        L = max(0,alpha_j_old + alpha_i_old - self.C)
                        H = min(self.C,alpha_j_old + alpha_i_old)
                    if L == H:
                        print("L=H")
                        continue
                    eta = 2 * self.Kernal(i,j)-self.Kernal(i,i)-self.Kernal(j,j)
                    if eta>= 0:
                        print("eta>=0")
                        continue
                    alpha_j_new_unc = alpha_j_old - y[j]*(E_i-E_j)/eta
                    self.alpha[j] = np.clip(alpha_j_new_unc,L,H)
                    if abs(self.alpha[j]-alpha_j_old) < 0.00001:
                        print("j not moving enough")
                        continue
                    self.alpha[i] += self.y[i]*self.y[j]*(alpha_j_old-self.alpha[j])
                    b_i_new = self.b - E_i -y[i]*self.Kernal(i,i)*(self.alpha[i]-alpha_i_old) \
                    -y[j]*self.Kernal(j,i)*(self.alpha[j]-alpha_j_old)
                    b_j_new = self.b - E_j -y[i]*self.Kernal(i,j)*(self.alpha[i]-alpha_i_old) \
                    -y[j]*self.Kernal(j,j)*(self.alpha[j]-alpha_j_old)
                    if (self.alpha[i]>0 and self.alpha[i]<self.C):
                        self.b = b_i_new
                    elif (self.alpha[j]>0 and self.alpha[j]<self.C):
                        self.b = b_j_new                    
                    else:
                        self.b = (b_i_new + b_j_new)/2
                    alphaPairsChanged += 1
                    print("External loop: %d; Internal loop i :%d; alphaPairsChanged :%d" % (iter,i,alphaPairsChanged))
            if (alphaPairsChanged == 0):
                iter += 1
            else:
                iter = 0
            print("Iteration number : %d" % iter)     
```

```python
def PlotSVM(X,y,classifier):
    X_positive = X[y==1]
    X_negative = X[y==-1]
    plt.figure(figsize = (6,4),dpi=120)
    plt.scatter(X_positive[:,0],X_positive[:,1],color = 'orange',s=30, alpha=0.7)
    plt.scatter(X_negative[:,0],X_negative[:,1],color = 'firebrick',s=30, alpha=0.7)
    W = np.sum(classifier.alpha * classifier.y.reshape(100,1) * classifier.X,axis = 0 )
    x1, x2 = np.min(X[:,0]), np.max(X[:,0])
    y1, y2 = float((-classifier.b- W[0]*x1)/W[1]), float((-classifier.b - W[0]*x2)/W[1])
    plt.plot([x1,x2],[y1,y2],color='darkmagenta')
    #绘制支持向量
    for i, alpha in enumerate(classifier.alpha):
        if abs(alpha) > 0:
            x_sv, y_sv = X[i]
            plt.scatter(x_sv, y_sv, s=150, c='none', alpha=0.7, linewidth=1.5, edgecolor='green')
    plt.title("Simple SMO")
    plt.show()
```
